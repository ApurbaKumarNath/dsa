```python
class MinHeap:
  def __init__(self, capacity):
    self.__heap = np.zeros(capacity + 1, dtype = None)
    self.__next = 1

  def insert(self, elem):
    if self.__next >= len(self.__heap):
      print('Heap Full')
    else:
      self.__heap[self.__next] = elem
      self.__next += 1
      self.swim()

  def swim(self):
    idx = self.__next - 1
    parent = idx // 2
    while idx > 1 and self.__heap[idx] < self.__heap[parent]:

      temp = self.__heap[parent]
      self.__heap[parent] = self.__heap[idx]
      self.__heap[idx] = temp

      idx = parent
      parent = idx // 2

  def extractMin(self):
    if self.__next == 1:
      print('Heap Empty')
    else:
      poppedVal = self.__heap[1]
      self.__heap[1] = self.__heap[self.__next - 1]
      self.__heap[self.__next - 1] = None
      self.__next -= 1
      self.sink()
      return poppedVal

  def sink(self):
    idx = 1
    child1 = idx * 2
    child2 = idx * 2 + 1
    while idx < self.__next:
      if child1 >= self.__next:
        break
      elif child2 >= self.__next:
        smaller = child1
      else:
        if self.__heap[child1] < self.__heap[child2]:
          smaller = child1
        else:
          smaller = child2

      if self.__heap[idx] > self.__heap[smaller]:

        temp = self.__heap[idx]
        self.__heap[idx] = self.__heap[smaller]
        self.__heap[smaller] = temp

        idx = smaller
        child1 = idx * 2
        child2 = idx * 2 + 1
      else:
        break

  def sort(self):
    temp = np.zeros(self.__next, dtype = None)
    for i in range(self.__next - 1, 0, -1):
      temp[i] = self.extractMin()

    return temp[1:]

  def printHeap(self):
    print(self.__heap[1 : self.__next])

  def getHeap(self):
    return self.__heap[1 : self.__next]
```

```python
# driver code
heap = MinHeap(10)

arr = np.array([11, 8, 10, 12, 7])
for i in arr:
  heap.insert(i)

print('Min heap with values inserted using insert():')
heap.printHeap()

print()
print('Min value removed using extractMin():')
print(heap.extractMin())

print()
print('Min heap after extractMin():')
heap.printHeap()

print()
print('Sorted Min heap using sort():')
print(heap.sort())
```

```python
class MaxHeap:
  def __init__(self, capacity):
    self.__heap = np.zeros(capacity + 1, dtype = None)
    self.__next = 1

  def insert(self, elem):
    if self.__next >= len(self.__heap):
      print('Heap Full')
    else:
      self.__heap[self.__next] = elem
      self.__next += 1
      self.swim()

  def swim(self):
    idx = self.__next - 1
    parent = idx // 2
    while idx > 1 and self.__heap[idx] > self.__heap[parent]:

      temp = self.__heap[parent]
      self.__heap[parent] = self.__heap[idx]
      self.__heap[idx] = temp

      idx = parent
      parent = idx // 2

  def extractMax(self):
    if self.__next == 1:
      print('Heap Empty')
    else:
      poppedVal = self.__heap[1]
      self.__heap[1] = self.__heap[self.__next - 1]
      self.__heap[self.__next - 1] = None
      self.__next -= 1
      self.sink()
      return poppedVal

  def sink(self):
    idx = 1
    child1 = idx * 2
    child2 = idx * 2 + 1

    while idx < self.__next:
      if child1 >= self.__next:
        break
      elif child2 >= self.__next:
        greater = child1
      else:
        if self.__heap[child1] > self.__heap[child2]:
          greater = child1
        else:
          greater = child2

      if self.__heap[idx] < self.__heap[greater]:

        temp = self.__heap[idx]
        self.__heap[idx] = self.__heap[greater]
        self.__heap[greater] = temp

        idx = greater
        child1 = idx * 2
        child2 = idx * 2 + 1
      else:
        break

  def sort(self):
    temp = np.zeros(self.__next, dtype = None)
    for i in range(self.__next - 1, 0, -1):
      temp[i] = self.extractMax()

    return temp[1:]

  def printHeap(self):
    print(self.__heap[1 : self.__next])

  def getHeap(self):
    return self.__heap[1 : self.__next]
```

```python
# driver code
heap = MaxHeap(10)

arr = np.array([12, 8, 10, 11, 7])
for i in arr:
  heap.insert(i)

print('Max heap with values inserted using insert():')
heap.printHeap()

print()
print('Max value removed using extractMax():')
print(heap.extractMax())

print()
print('Max heap after extractMax():')
heap.printHeap()

print()
print('Sorted Max heap using sort():')
print(heap.sort())
```

```python
def distributeTasks(tasks, m):
  heap = MinHeap(m)
  for i in range(m):
    heap.insert(0)

  for time in tasks:
    heap.insert(time + heap.extractMin())

  return heap.getHeap()

tasks = [2, 4, 7, 1, 6]
m = 4

print(distributeTasks(tasks, m))
```

```python
def kLargestElements(nums, k):
  heap = MaxHeap(len(nums))
  new = np.zeros(k, dtype = int)

  for num in nums:
    heap.insert(num)

  for i in range(k):
    new[i] = heap.extractMax()

  return new

nums = [4, 10, 2, 8, 6, 7]
k = 3

print(kLargestElements(nums, k))
```
