```python
# task 2
def bubbleSort(arr): # bubble sort algorithm (I tweaked it to stop the loop if no swaps are made)
    for i in range(len(arr)-1):
        swapped = False
        for j in range(len(arr)-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                swapped = True
        if not swapped:
            break

    return arr

with open('output2.txt', 'w') as output_file:
    with open('input2.txt', 'r') as input_file:

        line_1 = input_file.readline()
        arr = []
        for num in input_file.readline().split():
            arr.append(int(num))
        
        sorted_arr = bubbleSort(arr)
        for num in sorted_arr:
            output_file.write(f'{num} ')


# I have achieved the θ(n) for the best-case scenario after sorting the array, which means that as I have a sorted array now, if I sort it again (from the output_file),
# I don't need to swap any elements in the array. So, after one iteration, the algorithm will break out of the loop (check my code for that).
#  overall iteration will be n times only. This is the best-case scenario => time complexity = θ(n).
```

```python
# task 3
def mark_based_selection_sort(arr): # descending order
    for i in range(len(arr) - 1):
        min_idx = i
        min_changed = False

        for j in range(i + 1, len(arr)):
            if arr[j][1] > arr[min_idx][1]: # opposite of the normal selection sort
                min_idx = j
                min_changed = True
        
        if min_changed:
            arr[i], arr[min_idx] = arr[min_idx], arr[i]

    return arr

def id_based_selection_sort(arr):
    for i in range(len(arr) - 1):
        min_id_idx = i
        min_changed = False

        for j in range(i + 1, len(arr)):
            if arr[j][1] == arr[min_id_idx][1]: # if the marks are the same, then sort based on the id

                if arr[j][0] < arr[min_id_idx][0]: # min id will be placed first
                    min_id_idx = j
                    min_changed = True
        
        if min_changed:
            arr[i], arr[min_id_idx] = arr[min_id_idx], arr[i]

    return arr

with open('output3.txt', 'w') as output_file:
    with open('input3.txt', 'r') as input_file:

        length = int(input_file.readline())
        student_id = input_file.readline().split()
        student_marks = input_file.readline().split()

        arr = []
        for i in range(length):
            arr.append((int(student_id[i]), int(student_marks[i]))) # create a list of tuples --> (id, mark)

        sorted_arr = mark_based_selection_sort(arr)
        sorted_arr = id_based_selection_sort(sorted_arr)

        for student in sorted_arr:
            output_file.write(f'ID: {student[0]} Mark: {student[1]}\n')
```

```python
# 1st person's code
##################### Task 4  #####################
####################################################

# Insertion Sort        [We use this sort for Laarge/Multiple Data]
def insertionSort(schedule):
  for i in range(1, len(schedule)):
    train, dest, time = schedule[i]
    hour = int(time.split(":")[0])
    j = i - 1
    prevTrain, prevDest, prevTime = schedule[j][0], schedule[j][1], schedule[j][2]
    prevHour = int(prevTime.split(":")[0])
    while j>=0 and (prevTrain > train or (prevTrain == train and prevHour < hour)):
      schedule[j+1] = schedule[j]
      j -= 1
      prevTrain, prevDest, prevTime = schedule[j][0], schedule[j][1], schedule[j][2]
      prevHour = int(prevTime.split(":")[0])
    schedule[j+1] = (train, dest, time)
  return schedule

# Read the file
inp4 = open("input4.txt","r")
# Generate a output file by running the code
out4 = open("output4.txt","w")

total = int(inp4.readline())

schedule = []

for i in range(total):
  info = inp4.readline().split(" ")
  name = info[0]
  district = info[-3]
  time = info[-1][:-1]
  schedule.append([name,district,time])

sortedSchedule = insertionSort(schedule)

for i in range(total):
  out4.write(f"{sortedSchedule[i][0]} will departure for {sortedSchedule[i][1]} at {sortedSchedule[i][2]}\n")

# must close the file; Otherwise the output will not Update/generated
out4.close()


###Explanation
# Here I have used insertion sort to handle large number of data.The insertionSort function iterates through the schedule and compares each train's departure time and name with the previous ones.
#It rearranges the elements in a way that ensures the schedule is sorted based on departure time and, in case of a tie, by train name.
# The main loop extracts info by spliting and sends them to schedule list. Then schedule is sorted using insertion sort and output
# is written in designated file.
```
