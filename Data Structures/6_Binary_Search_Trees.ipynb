##Tree data structure (From Codebasics)

```python
class TreeNode:
  def __init__(self, data):
    self.data = data
    self.children = []
    self.parent = None

  def addChild(self, child):
    child.parent = self
    self.children.append(child)

  def getLevel(self):
    level = 0
    p = self.parent
    while p:
      level += 1
      p = p.parent
    return level

  def printTree(self):
    if self.parent:
      prefix = ' ' * self.getLevel() * 3 + '|__'
    else:
      prefix = ''

    print(prefix + self.data)
    for child in self.children:
      child.printTree()
```

## Binary Search Tree (From Codebasics)

```python
class BinarySearchTreeNode:
  def __init__(self, data):
    self.data = data
    self.left = None
    self.right = None

  def addChild(self, data):
    if self.data == data:
      return

    if data < self.data:
      if self.left:
        self.left.addChild(data)
      else:
        self.left = BinarySearchTreeNode(data)

    else:
      if self.right:
        self.right.addChild(data)
      else:
        self.right = BinarySearchTreeNode(data)

  def search(self, val):
    if self.data == val:
      return True

    if val < self.data:
      if self.left:
        return self.left.search(val)
      else:
        return False

    if val > self.data:
      if self.right:
        return self.right.search(val)
      else:
        return False

  def inOrderTraversal(self):
    elements = []

    if self.left:
      elements += self.left.inOrderTraversal()

    elements.append(self.data)

    if self.right:
      elements += self.right.inOrderTraversal()

    return elements

  def preOrderTraversal(self):
    elements = []

    elements.append(self.data)

    if self.left:
      elements += self.left.preOrderTraversal()

    if self.right:
      elements += self.right.preOrderTraversal()

    return elements

  def postOrderTraversal(self):
    elements = []

    if self.left:
      elements += self.left.postOrderTraversal()

    if self.right:
      elements += self.right.postOrderTraversal()

    elements.append(self.data)

    return elements

  def calculateSum(self):
    sum = 0

    if self.left:
      sum += self.left.calculateSum()

    sum += self.data

    if self.right:
      sum += self.right.calculateSum()

    return sum

  def findMin(self):
    if self.left:
      return self.left.findMin()
    return self.data

  def findMax(self):
    if self.right:
      return self.right.findMax()
    return self.data

  # 2 ways to delete: inorder successor & inorder predecessor
  # here, using inorder successor
  def delete(self, val):
    if val < self.data:
      if self.left:
        self.left = self.left.delete(val)

    elif val > self.data:
      if self.right:
        self.right = self.right.delete(val)

    else:
      if self.left == None and self.right == None:
        return None

      if self.left == None:
        return self.right

      if self.right == None:
        return self.left

      min_val = self.right.findMin()
      self.data = min_val
      self.right = self.right.delete(min_val)

    return self
```

```python
# function to get the sum of nodes
def sumOfNodes(root):
  if root == None:
    return 0
  # sum(root) = root.elem + sum(left_subtree_of_root) + sum(right_subtree_of_root)
  return root.elem + sumOfNodes(root.left) + sumOfNodes(root.right)
```

```python
# height = longest distance from node to leaf
def height(node):
  if node == None:
    return -1
  # return 1 + max(height(node.left), height(node.right)) # or,

  left_sum = height(node.left)
  right_sum = height(node.right)

  if left_sum > right_sum:
    return 1 + left_sum
  else:
    return 1 + right_sum

root = tree_construction([None, 1,2,3,4,6,4,7,3])
print(sumOfNodes(root))
print(height(root))
```

```python
# depth or level = distance from node to root

def depth(root, key, level = 0): # key = given node.elem
  # traverse from root. return the level at which key == root.elem

  if root == None:
    return -1 # returning something that cannot be a level. So, -1 here means "Not found"

  if root.elem == key:
    return level

  left = depth(root.left, key, level + 1)

  right = depth(root.right, key, level + 1)

  if left > right:  # either right is -1 or left is -1. Greater one is the answer.
    return left
  else:
    return right

print(depth(root, 7))
```

```python
# Do it yourself
# Check if a Tree is Complete Binary Tree:
# hint = every level, except possibly the last, is completely filled, and all nodes are as far left as possible.

class BT:
  def __init__(self, arr):
    self.root = self.tree_construction(arr)
    self.found_level = -1
    self.finisher = False

  def tree_construction(self, arr, i = 1):
    if i>=len(arr) or arr[i] == None:
      return None
    p = BTNode(arr[i])
    p.left = tree_construction(arr, 2*i)
    p.right = tree_construction(arr, 2*i+1)
    return p

  def if_complete(self, root, level = 0):
    if root == None:
      if level == self.found_level:
        self.finisher = True
      return

    elif root.left == None and root.right == None:
      if self.found_level == -1:
        self.found_level = level
      else:
        if level != self.found_level:
          return False
        else:
          if self.finisher == True:
            return False

    elif root.left != None and root.right == None:
      if level == self.found_level:
        return False
      elif level + 1 != self.found_level:
        return False


    elif root.left == None and root.right != None:
      return False

    elif root.left != None and root.right != None:
      if level == self.found_level:
        return False

    left = self.if_complete(root.left, level + 1)
    right = self.if_complete(root.right, level + 1)

    if left == False or right == False:
      return False
    else:
      return True

tree_object = BT([None, 1,2,3,None,None,4,7])
#tree_object.inOrder()
#print()
print(tree_object.if_complete(tree_object.root))

# Do it using a Queue (geeksforgeeks)
```

```python
class BTNode:
    def __init__(self, elem):
        self.elem = elem
        self.right = None
        self.left = None

def is_symmetric(root):
    """
    Checks if a binary tree is symmetric (a mirror image of itself).

    Args:
      root: The root node of the binary tree.

    Returns:
      True if the tree is symmetric, False otherwise.
    """
    if root is None:
        return True

    return are_mirrors(root.left, root.right)

def are_mirrors(node1, node2):
    """
    Helper function to check if two subtrees are mirror images of each other.

    Args:
        node1: The root of the first subtree.
        node2: The root of the second subtree.

    Returns:
        True if the subtrees are mirror images, False otherwise.
    """
    if node1 is None and node2 is None:
        return True
    if node1 is None or node2 is None:
        return False
    if node1.elem != node2.elem:
        return False
    return are_mirrors(node1.left, node2.right) and are_mirrors(node1.right, node2.left)

# Example usage (creating a symmetric tree):
root = BTNode(1)
root.left = BTNode(2)
root.right = BTNode(2)
root.left.left = BTNode(3)
root.left.right = BTNode(4)
root.right.left = BTNode(4)
root.right.right = BTNode(3)

print(f"Is the tree symmetric? {is_symmetric(root)}")  # Output: True

# Example usage (creating a non-symmetric tree):
root2 = BTNode(1)
root2.left = BTNode(2)
root2.right = BTNode(3)

print(f"Is the tree symmetric? {is_symmetric(root2)}")  # Output: False
```

```python
# Convert a sorted (unbalanced) BST to a Balanced BST
# Python program to convert a left unbalanced
# BST to a balanced BST

class Node:
    def __init__(self, value):
        self.data = value
        self.left = None
        self.right = None

# Inorder traversal to store elements of the
# tree in sorted order
def store_inorder(root, nodes):
    if root is None:
        return

    # Traverse the left subtree
    store_inorder(root.left, nodes)

    # Store the node data
    nodes.append(root.data)

    # Traverse the right subtree
    store_inorder(root.right, nodes)

# Function to build a balanced BST from a sorted array
def build_balanced_tree(nodes, start, end):

    # Base case
    if start > end:
        return None

    # Get the middle element and make it the root
    mid = (start + end) // 2
    root = Node(nodes[mid])

    # Recursively build the left and right subtrees
    root.left = build_balanced_tree(nodes, start, mid - 1)
    root.right = build_balanced_tree(nodes, mid + 1, end)

    return root

# Function to balance a BST
def balance_bst(root):
    nodes = []

    # Store the nodes in sorted order
    store_inorder(root, nodes)

    # Build the balanced tree from the sorted nodes
    return build_balanced_tree(nodes, 0, len(nodes) - 1)

# Function to print the tree (Inorder Traversal)
def inorder(root):
    if root is None:
        return

    inorder(root.left)
    print(root.data, end=' ')
    inorder(root.right)

if __name__ == "__main__":

    # Constructing an unbalanced BST
    #        10
    #       /  \
    #      5    15
    #     /       \
    #    2         20
    #   /
    #  1

    root = Node(10)
    root.left = Node(5)
    root.left.left = Node(2)
    root.left.left.left = Node(1)
    root.right = Node(15)
    root.right.right = Node(20)

    balanced_root = balance_bst(root)
    inorder(balanced_root)
    print()
```

```python
# Insert operation
def insert(root, elem):
  if root == None:
    return BTNode(elem)

  if root.elem > elem:
    root.left = insert(root.left, elem)

  else:
    root.right = insert(root.right, elem)

  return root # to make sure root.left doesn't change if root != None.
```
